```python
import numpy as np
import pandas as pd

data = pd.read_excel (r'C:\Users\aliul\Desktop\WORKING FILE FOR REGRESSION ANALYSIS.xlsx')

data.head()
```

```python
data1 = pd.DataFrame(data)

data1.set_index(pd.Index(range(1,101)))
```

```python
x = data[['TOTAL DISTANCE','TOTAL UNITS FLOW','TOTAL PER UNIT COST']]
y = data[['TOTAL COST']]
print(x,y)
```

```python
from sklearn.model_selection import train_test_split as tts

x_train , x_test , y_train , y_test = tts(x,y)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(x_train,y_train)

y_hat = model.predict(x_test)
```

```python

l1=pd.DataFrame(y_hat.astype(int))
l2=pd.DataFrame(np.asanyarray(y_test))
l3=l1-l2
l1.columns = ['predicted']
l2.columns = ['actual']
l3.columns = ['difference']
l4 = l1.join(l2)
l5=l4.join(l3)
l5
           
```

```python
from sklearn import metrics

metrics.mean_absolute_error(y_test,y_hat)
```
